 P166: Remove suffix and prefix and return remaining string.

```python
# txt = input()
txt = 'abcdjklcba'
result = []
for i in range(len(txt)):
    if txt[i] != txt[-(i+1)]:
        result.append(txt[i:-i])
        break
if result != []:
    print(*result)
else:
    print('Minimal string is empty')
```

 P170: Count number of jumping number (diff = 1)

```python
# N = int(input())
N = 40
if N > 9:
    count = 10
    for i in range(10, N+1):
        num = list(map(int, list(str(i))))
        if abs(num[0]-num[1]) == 1:
            # print(num)
            count += 1
    print(count)
else:
    print(N)
```
